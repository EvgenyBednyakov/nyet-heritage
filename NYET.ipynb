```python
%matplotlib inline

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import matplotlib
```

```python
df = pd.read_csv ('NYET(1).csv') #Читаем CSV
```

```python
df2 = []
df2 = df[['date', 'print_name', 'text', 'type']]
```

```python
df2.date = pd.to_datetime(df2.date,unit='s')
```

matplotlib.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (18, 9)

df3 = df2[(df2.date > '2016-01-01') & (df2.date < '2016-06-01')]

per = df3.date.dt.to_period("d") 
df3=df3.groupby(per)
df3.size().plot()

plt.ylabel('message amount')

```python
df3 = df2[(df2.date > '2016-01-01') & (df2.date < '2016-01-30')]
per = df3.date.dt.to_period("d")
names = df3.print_name.unique()

df3.groupby(per).size().plot(linewidth=6)

for x in names: 
        df3[df3.print_name == x].groupby(per).size().plot()
        
```
